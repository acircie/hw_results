```python
import numpy as np
import matplotlib.pyplot as plt
```

#### HW9 results

```python
tau_star = np.linspace(10, 0)
epsilon = 0.3

tau =  tau_star/np.sqrt(3. * epsilon)

plt.plot(tau, 1. - ((1. - np.sqrt(epsilon))/(1. + np.sqrt(epsilon))) * np.exp(-tau), label = '$I^{(+)}/B$')

plt.plot(tau, 1. - np.exp(-tau), label = '$I^{(-)}/B$')

plt.xlim([10, 0])
plt.ylim([0,1.1])
plt.xlabel('$\\tau$', fontsize = 16)

plt.legend(fontsize = 16)

title_str = f'$\epsilon$ = {str(epsilon)}'
plt.title(title_str, fontsize = 16)

plt.tick_params(which = 'both', direction = 'in', top = True, right = True,
               grid_alpha = 0.5, grid_linestyle = '--', labelsize = 10)
plt.grid(True)
plt.show()
```

```python
plt.plot(tau, 1. - (1. - np.sqrt(epsilon)) * np.exp(-tau), label = 'S/B')
plt.plot(tau, 1. - 1. / (1. + np.sqrt(epsilon)/(1. + np.sqrt(epsilon))) * np.exp(-tau), label = 'H/B')
plt.plot(tau, np.exp(-tau), label = 'F/F(0)')

plt.xlim([10, 0])
plt.ylim([0,1.1])
plt.xlabel('$\\tau$', fontsize = 16)

plt.legend(fontsize = 16)

title_str = f'$\epsilon$ = {str(epsilon)}'
plt.title(title_str, fontsize = 16)

plt.tick_params(which = 'both', direction = 'in', top = True, right = True,
               grid_alpha = 0.5, grid_linestyle = '--', labelsize = 10)
plt.grid(True)
plt.show()
```

#### HW11 results

```python
tau_star = np.linspace(10, 0)
epsilon = 0.3

tau =  tau_star/np.sqrt(3. * epsilon)

tau_star_0 = 10.
tau_0 = tau_star_0/np.sqrt(3*epsilon)

# factors to define I+ and I-
eps_1 = 1 + np.sqrt(epsilon)
eps_2 = 1 - np.sqrt(epsilon)

# numerator of the fraction in I+
num_1 = eps_1 * np.exp(tau_star - tau_star_0/2) + eps_2 * np.exp(-tau_star + tau_star_0/2)
# numerator of the fraction in I-
num_2 = eps_2 * np.exp(tau_star - tau_star_0/2) + eps_1 * np.exp(-tau_star + tau_star_0/2)
# denominator
den = np.sqrt(epsilon) * np.sinh(tau_star_0/2) + np.cosh(tau_star_0/2)

# Define I+/B
I1 = 1 - 0.5*(num_1/den)

# Define I-/B
I2 = 1 - 0.5*(num_2/den)

# plot the results
fig, ax = plt.subplots()

ax.plot(tau, I1, label = '$I^{(+)}/B$')
ax.plot(tau, I2, label = '$I^{(-)}/B$')

ax.set_xlim([tau_0, 0])
ax.set_ylim([0,1.1])
ax.set_xlabel('$\\tau$', fontsize = 16)

ax.legend(fontsize = 16)

title_str = f'$\epsilon$ = {str(epsilon)}'
ax.set_title(title_str, fontsize = 16)

ax.tick_params(which = 'both', direction = 'in', top = True, right = True,
               grid_alpha = 0.5, grid_linestyle = '--', labelsize = 10)
ax.grid(True)
```

```python
# Define F/F(0)
f = (np.exp(-tau_star + tau_star_0/2) - np.exp(tau_star - tau_star_0/2))/(2*np.sinh(tau_star_0/2))

# plot the results
fig, ax = plt.subplots()

ax.plot(tau, f, label = 'F/F(0)')

ax.set_xlim([tau_0, 0])
ax.set_xlabel('$\\tau$', fontsize = 16)

ax.legend(fontsize = 16)

title_str = f'$\epsilon$ = {str(epsilon)}'
ax.set_title(title_str, fontsize = 16)

ax.tick_params(which = 'both', direction = 'in', top = True, right = True,
               grid_alpha = 0.5, grid_linestyle = '--', labelsize = 10)
ax.grid(True)
```
